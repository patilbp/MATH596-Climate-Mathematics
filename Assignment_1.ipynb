```python
############### PROBLEM - 2.7 (b) ################
############### LINEAR TREND LINES : TMIN ########

import os, pandas as pd, numpy as np, matplotlib.pyplot as plt, seaborn as sns
from sklearn.linear_model import LinearRegression
os.chdir("/Users/bhagy/Downloads/data/data")

file = "CA042239T.csv"
col_list=['YEAR  ','Month  ','TMIN (F) ']

df = pd.read_csv(file, sep=',',usecols=col_list)
df = df[(df['YEAR  ']>=1951) & (df['YEAR  ']<=2010) & (df['Month  ']==1)]

timeperiod_1 = df['YEAR  '].to_numpy()
tmin_1 = pd.to_numeric(df['TMIN (F) '])
tmin_1 = tmin_1.to_numpy()


plt.figure(figsize=(25,10))
plt.plot(timeperiod_1,tmin_1,'black')

m, b = np.polyfit(timeperiod_1,tmin_1,1)

def f(x):
    return m*x + b

plt.plot(timeperiod_1,f(timeperiod_1),'-r',label='1951-2010')

filter_1 = df.where((df['YEAR  ']>=1961) & (df['YEAR  ']<=2010))
filter_1=filter_1.dropna()

timeperiod_2 = filter_1['YEAR  ']
tmin_2 = filter_1['TMIN (F) ']
tmin_2 = pd.to_numeric(tmin_2)
tmin_2 = tmin_2.to_numpy()

m, b = np.polyfit(timeperiod_2,tmin_2,1)


plt.plot(timeperiod_2,f(timeperiod_2),'-b',label='1961-2010')

filter_2 = df.where((df['YEAR  ']>=1971) & (df['YEAR  ']<=2010))
filter_2=filter_2.dropna()

timeperiod_3 = filter_2['YEAR  ']
tmin_3 = filter_2['TMIN (F) ']
tmin_3 = pd.to_numeric(tmin_3)
tmin_3 = tmin_3.to_numpy()

m, b = np.polyfit(timeperiod_3,tmin_3,1)


plt.plot(timeperiod_3,f(timeperiod_3),'-g',label='1971-2010')

filter_3 = df.where((df['YEAR  ']>=1981) & (df['YEAR  ']<=2010))
filter_3=filter_3.dropna()

timeperiod_4 = filter_3['YEAR  ']
tmin_4 = filter_3['TMIN (F) ']
tmin_4 = pd.to_numeric(tmin_4)
tmin_4 = tmin_4.to_numpy()

m, b = np.polyfit(timeperiod_4,tmin_4,1)


plt.plot(timeperiod_4,f(timeperiod_4),'brown',label='1981-2010')
plt.title('Linear Trend Lines : Tmin')
plt.xlabel('Year')
plt.ylabel('TMIN (F)')
```
